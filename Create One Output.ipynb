```python
import pandas as pd, numpy as np, time, datetime, csv, os
```

```python
# specify agency
list_of_agencies = list(os.walk('gtfs'))[0][1]
```

```python
# Export all bus stops
data = []
for agency in list_of_agencies:
    agency_name = pd.read_csv("gtfs/" + agency + '/agency.txt')['agency_name'][0]
    agency_name = agency_name.replace("/","-")
    filename = "output/%s_test.csv" % agency_name
    try:
        data.append(pd.read_csv(filename))
    except FileNotFoundError:
        print(filename)
        pass
df = pd.concat(data)
df.to_csv("output/AllBus.csv")
```

```python
# Export all rail stops
data = []
for agency in list_of_agencies:
    agency_name = pd.read_csv("gtfs/" + agency + '/agency.txt')['agency_name'][0]
    agency_name = agency_name.replace("/","-")
    filename = "output/%s_RAIL_test.csv" % agency_name
    try:
        data.append(pd.read_csv(filename))
    except FileNotFoundError:
        print(filename)
        pass
df_rail = pd.concat(data)
df_rail.to_csv("output/AllRail.csv")
```
